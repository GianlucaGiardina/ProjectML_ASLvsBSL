```python

import tensorflow as tf
from tensorflow.keras.applications import DenseNet121, MobileNetV2
from tensorflow.keras.applications.densenet import preprocess_input as preprocess_densenet
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as preprocess_mobilenet
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Imposta i percorsi ai modelli
densenet_model_path =  "/tmp/Densenet_model.h5"
# Sostituisci con il tuo percorso
mobilenet_model_path =  "/tmp/MobileNetV2Dataset2_model.h5"
# Carica i modelli
densenet_model = tf.keras.models.load_model(densenet_model_path)
mobilenet_model = tf.keras.models.load_model(mobilenet_model_path)

# Lista delle etichette delle classi
class_labels_list = [
    'A_ASL', 'A_BSL', 'B_ASL', 'B_BSL', 'C_ASL', 'C_BSL', 'D_ASL', 'D_BSL',
    'E_ASL', 'E_BSL', 'F_ASL', 'F_BSL', 'G_ASL', 'G_BSL', 'H_ASL', 'H_BSL',
    'I_ASL', 'I_BSL', 'K_ASL', 'K_BSL', 'L_ASL', 'L_BSL', 'M_ASL', 'M_BSL',
    'N_ASL', 'N_BSL', 'O_ASL', 'O_BSL', 'P_ASL', 'P_BSL', 'Q_ASL', 'Q_BSL',
    'R_ASL', 'R_BSL', 'S_ASL', 'S_BSL', 'T_ASL', 'T_BSL', 'U_ASL', 'U_BSL',
    'V_ASL', 'V_BSL', 'W_ASL', 'W_BSL', 'X_ASL', 'X_BSL', 'Y_ASL', 'Y_BSL'
]

# Dizionario per mappare gli indici delle classi alle etichette
class_labels = {i: label for i, label in enumerate(class_labels_list)}


# Funzione per l'upload dell'immagine
def upload_image():
    uploaded = files.upload()
    for filename in uploaded.keys():
        img_path = '/content/' + filename
        return img_path

# Funzione per caricare e pre-processare l'immagine
def load_and_preprocess_image(img_path, model_name):
    if model_name == 'densenet':
        a=b=80
    elif model_name == 'mobilenet':
        a=b=160
    img = image.load_img(img_path, target_size=(a, b))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    if model_name == 'densenet':
        return preprocess_densenet(img_array)
    elif model_name == 'mobilenet':
        return preprocess_mobilenet(img_array)

# Funzione per ottenere le predizioni
def get_predictions(model, img_array, labels):
    predictions = model.predict(img_array)
    top_indices = predictions[0].argsort()[-5:][::-1]  # Considera solo i primi 5 risultati
    results = [(labels[i], predictions[0][i]) for i in top_indices]
    return results
# Carica i label di ImageNet
labels_path = tf.keras.utils.get_file('ImageNetLabels.txt', 'https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')
with open(labels_path) as f:
    labels = f.read().splitlines()


# Funzione per visualizzare i risultati con valori > 0
def plot_predictions(img_path, densenet_results, mobilenet_results):
    img = image.load_img(img_path, target_size=(224, 224))
    plt.figure(figsize=(10, 5))

    # Plot per DenseNet
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title('DenseNet Predictions')

    # Filtra e plotta solo i valori > 0 per DenseNet
    densenet_results_filtered = [(label, score) for label, score in densenet_results if score > 0]
    plt.text(0, -10, '\n'.join([f"{res[0]}: {res[1]:.2f}" for res in densenet_results_filtered]), fontsize=12)

    # Plot per MobileNetV2
    plt.subplot(1, 2, 2)
    plt.imshow(img)
    plt.axis('off')
    plt.title('MobileNetV2 Predictions')

    # Filtra e plotta solo i valori > 0 per MobileNetV2
    mobilenet_results_filtered = [(label, score) for label, score in mobilenet_results if score > 0]
    plt.text(0, -10, '\n'.join([f"{res[0]}: {res[1]:.2f}" for res in mobilenet_results_filtered]), fontsize=12)

    plt.show()

# Funzione per eseguire tutto il processo
def run_prediction():
    img_path = upload_image()

    if img_path:
        img_densenet = load_and_preprocess_image(img_path, 'densenet')
        img_mobilenet = load_and_preprocess_image(img_path, 'mobilenet')

        densenet_predictions = get_predictions(densenet_model, img_densenet, class_labels)
        mobilenet_predictions = get_predictions(mobilenet_model, img_mobilenet, class_labels)

        plot_predictions(img_path, densenet_predictions, mobilenet_predictions)
    else:
        print("Nessun file selezionato.")

# Funzione per eseguire tutto il processo
def run_prediction():
    img_path = upload_image()

    if img_path:
        img_densenet = load_and_preprocess_image(img_path, 'densenet')
        img_mobilenet = load_and_preprocess_image(img_path, 'mobilenet')

        densenet_predictions = get_predictions(densenet_model, img_densenet, class_labels)
        mobilenet_predictions = get_predictions(mobilenet_model, img_mobilenet, class_labels)

        plot_predictions(img_path, densenet_predictions, mobilenet_predictions)
    else:
        print("Nessun file selezionato.")
# Esegui l'applicazione
run_prediction()
```

```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from tensorflow.keras.preprocessing import image

# Widget per selezionare l'alfabeto
alphabet_dropdown = widgets.Dropdown(
    options=['ASL', 'BSL'],
    description='Alphabet:',
    disabled=False,
)

# Widget per inserire la parola
word_textbox = widgets.Text(
    value='',
    placeholder='Enter a word',
    description='Word:',
    disabled=False
)

# Lista delle etichette delle classi
class_labels_list = [
    'A_ASL', 'A_BSL', 'B_ASL', 'B_BSL', 'C_ASL', 'C_BSL', 'D_ASL', 'D_BSL',
    'E_ASL', 'E_BSL', 'F_ASL', 'F_BSL', 'G_ASL', 'G_BSL', 'H_ASL', 'H_BSL',
    'I_ASL', 'I_BSL', 'K_ASL', 'K_BSL', 'L_ASL', 'L_BSL', 'M_ASL', 'M_BSL',
    'N_ASL', 'N_BSL', 'O_ASL', 'O_BSL', 'P_ASL', 'P_BSL', 'Q_ASL', 'Q_BSL',
    'R_ASL', 'R_BSL', 'S_ASL', 'S_BSL', 'T_ASL', 'T_BSL', 'U_ASL', 'U_BSL',
    'V_ASL', 'V_BSL', 'W_ASL', 'W_BSL', 'X_ASL', 'X_BSL', 'Y_ASL', 'Y_BSL'
]

# Dizionario per mappare gli indici delle classi alle etichette
class_labels = {i: label for i, label in enumerate(class_labels_list)}

# Funzione per caricare le immagini dalla cartella locale alfabeto
def load_images_from_folder(alphabet, word):
    images = []
    alphabet_folder = f"/tmp/alfabeto"

    for char in word:
        img_name = f"{char}_{alphabet}.jpg"
        img_path = os.path.join(alphabet_folder, img_name)

        # Carica l'immagine e aggiungi all'array
        img = image.img_to_array(image.load_img(img_path, target_size=(224, 224)))
        images.append(img)

    return np.array(images)

# Funzione per eseguire le predizioni sui modelli
def run_predictions(images):
    # Carica i modelli (sostituisci con i tuoi percorsi)
    densenet_model_path = "/tmp/Densenet_model.h5"
    mobilenet_model_path = "/tmp/MobileNetV2Dataset2_model.h5"
    densenet_model = tf.keras.models.load_model(densenet_model_path)
    mobilenet_model = tf.keras.models.load_model(mobilenet_model_path)

    # Preprocessa le immagini per DenseNet
    images_preprocessed_densenet = tf.image.resize(images, (80, 80))
    images_preprocessed_densenet = tf.keras.applications.densenet.preprocess_input(images_preprocessed_densenet)

    # Preprocessa le immagini per MobileNetV2
    images_preprocessed_mobilenet = tf.image.resize(images, (160, 160))
    images_preprocessed_mobilenet = tf.keras.applications.mobilenet_v2.preprocess_input(images_preprocessed_mobilenet)

    # Ottieni le predizioni
    densenet_predictions = densenet_model.predict(images_preprocessed_densenet)
    mobilenet_predictions = mobilenet_model.predict(images_preprocessed_mobilenet)

    return densenet_predictions, mobilenet_predictions

# Funzione per visualizzare le immagini in fila con le predizioni dei modelli
def plot_images_with_predictions(images, densenet_predictions, mobilenet_predictions,word):
    num_images = len(images)
    plt.figure(figsize=(15, 8))

    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        plt.imshow(images[i] / 255.0)
        plt.axis('off')

        # Ottieni l'etichetta predetta per DenseNet
        densenet_predicted_label = class_labels[np.argmax(densenet_predictions[i])]
        # Ottieni l'etichetta predetta per MobileNetV2
        mobilenet_predicted_label = class_labels[np.argmax(mobilenet_predictions[i])]

        # Scrivi la lettera corrispondente sopra l'immagine in rosso
        plt.text(0.5, 1.15, word[i], color='red', fontsize=12, ha='center', transform=plt.gca().transAxes)

        plt.title(f'Real: {word[i]} \n DenseNet: {densenet_predicted_label}\nMobileNetV2: {mobilenet_predicted_label}')

    plt.tight_layout()
    plt.show()

# Funzione principale per eseguire l'applicazione
def main(alphabet, word):
    images = load_images_from_folder(alphabet, word)
    densenet_preds, mobilenet_preds = run_predictions(images)
    plot_images_with_predictions(images, densenet_preds, mobilenet_preds,word)

# Esegui l'applicazione
display(alphabet_dropdown, word_textbox)

run_button = widgets.Button(description="Run Prediction")
display(run_button)

output = widgets.Output()
display(output)

def on_run_button_clicked(b):
    with output:
        clear_output(wait=True)
        alphabet = alphabet_dropdown.value
        word = word_textbox.value.upper()
        main(alphabet, word)

run_button.on_click(on_run_button_clicked)
```
